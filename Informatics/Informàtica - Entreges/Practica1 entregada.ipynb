```python
import numpy as fa
import matplotlib.pylab as plt
x = fa.linspace(-fa.pi, fa.pi, 10)
x1 = fa.linspace(-fa.pi, fa.pi, 100)
y = fa.sin(x)
y+=fa.random.normal(0, 0.3, size=x.shape)
y1=fa.sin(x1)
plt.plot(x,y, 'bo')
plt.plot(x1,y1, 'g')
plt.show()
```
