##Pas1: definició d'una funció d'entrada de dades

Se'm demana que creii una funció amb el nom: `demana_enter` i amb els arguments: `(text,min,max)` amb un valor per defecte min=0 i max=9. Aquesta funció ha de demanar per teclat un valor enter. Si aquest valor enter no verifica que és ≥ min i ≤ max s'ha de tornar a demanar un altre enter. Quan es doni un valor en el rang acceptable, l'ha de retornar.

```python
def demana_enter (min,max):
    while True:
        try:
            enter = int(input("Escriu un nombre enter entre " + str(min) + " i " + str(max) + ": "))
            if enter >= min and enter <= max:
                print ("Valor correcte")
                break
            else:
                print ("Has escrit un enter, però no està en l'interval [" + str(min) + ", " + str(max) +"]!" )
        except ValueError:
            print ("Això no és un enter!")
    return(enter)
       
demana_enter(0,9)
```

## Pas2: inicialització de la finestra

* Un color de fons al meu gust

<span style="color:blue">Com que les explicacions de la creació d'una finestra i d'una tortuga es fan molt repetitives sobretot venint de fer la pràctica de definició de funcions per dibuixar els ninots amb turtle, no crec que sigui necessari tornar a escriure perquè escric el codi que escric.
```python
def ini_finestra():#creo la funció amb el nom que se m'ha demanat
    import turtle as t #importo la llibreria turtle
    finestra = t.Screen() #creo la finestra de treball
    finestra.title("Buscamines")#li determino un títol (el demanat)
    finestra.bgcolor("green")#li defineixo el color del fons al meu gust
    finestra.exitonclick()#li afageixo l'opció d'exitonclick
    return(finestra)#li faig retornar la finestra inicialitzada perquè s'usi en futures manipulacions
    
ini_finestra() #crido la funció
```

```python
#copio el codi del pas 2, excepte l'exitonclick, ja que vull que aquest quedi al final.
def ini_finestra():
    finestra = t.Screen()
    finestra.setup(600,600)
    finestra.title("Buscamines")
    finestra.bgcolor("green")
    return(finestra)

#copio el codi del pas 3
def ini_tortuga():
    Ripo= t.Turtle()
    Ripo.pensize("1")
    Ripo.speed(10)
    Ripo.color("black")
    return(Ripo)
            
def ini_graella (tortuga):
    for i in range(11):
        tortuga.penup()
        tortuga.goto(-250,-250)
        tortuga.setheading(90)
        tortuga.fd(50*i)
        tortuga.setheading(0)
        tortuga.pendown()
        tortuga.fd(500)
    for i in range (11):
        tortuga.penup()
        tortuga.goto(-250,-250)
        tortuga.setheading(0)
        tortuga.fd(50*i)
        tortuga.setheading(90)
        tortuga.pendown()
        tortuga.fd(500)
    return

import turtle as t #importo turtle a fora per la mateixa raó anterior (del pas3), perquè afecti a totes les funcions
myturtle = ini_tortuga() #crido la funció tortuga guardant el return amb la variable "myturtle"
mywindow = ini_finestra() #crido la funció finestra guardant el return amb la variable "mywindow"
ini_graella(myturtle) #faig servir la tortuga definida a la funció ini_tortuga per traçar la graella
mywindow.exitonclick() #Tanco la finestra definida a la funció ini_finestra, és a dir la finestra que ha retornat la funció ini_finestra i que ha estat guardada amb la variable "mywindow".
```

```python
def demana_enterx (min,max):
    while True:
        try:
            enterx = int(input("Escriu un nombre enter entre " + str(min) + " i " + str(max) + " per determinar la fila: "))
            if enterx >= min and enterx <= max:
                print ("El valor introduït és correcte. Has triat la fila número: " + str(enterx))
                return(enterx)
                break
            else:
                print ("Has escrit un enter, però no està en l'interval [" + str(min) + ", " + str(max) +"]!" )
        except ValueError:
            print ("Això no és un enter!")

def demana_entery (min,max):
    while True:
        try:
            entery = int(input("Escriu un nombre enter entre " + str(min) + " i " + str(max) + " per determinar la columna: "))
            if entery >= min and entery <= max:
                print ("El valor introduït és correcte. Has triat la columna número: " + str(entery))
                return(entery)
                break
            else:
                print ("Has escrit un enter, però no està en l'interval [" + str(min) + ", " + str(max) +"]!" )
        except ValueError:
            print ("Això no és un enter!")
      
def ini_finestra():
    finestra = t.Screen()
    finestra.setup(600,600)
    finestra.title("Buscamines")
    finestra.bgcolor("white")
    return(finestra)
    
def ini_tortuga():
    Ripo= t.Turtle("turtle")
    Ripo.pensize("1")
    Ripo.speed(50)
    Ripo.color("black")
    return(Ripo)
            
def ini_graella (tortuga):
    for i in range(11):
        tortuga.penup()
        tortuga.goto(-250,-250)
        tortuga.setheading(90)
        tortuga.fd(50*i)
        tortuga.setheading(0)
        tortuga.pendown()
        tortuga.fd(500)
    for i in range (11):
        tortuga.penup()
        tortuga.goto(-250,-250)
        tortuga.setheading(0)
        tortuga.fd(50*i)
        tortuga.setheading(90)
        tortuga.pendown()
        tortuga.fd(500)
    tortuga.penup()
    tortuga.fd(10)
    tortuga.rt(90)
    tortuga.bk(480)
    p=-1
    while p < 9:
        tortuga.write(p+1)
        tortuga.fd(50)
        p+=1
    tortuga.goto(-270,-235)
    tortuga.setheading(90)
    o=10
    while o>0:
        tortuga.write(o-1)
        tortuga.fd(50)
        o = o - 1
    tortuga.goto(-225,-270)
    tortuga.setheading(0)
    p=-1
    while p < 9:
        tortuga.write(p+1)
        tortuga.fd(50)
        p+=1
    tortuga.bk(10)
    tortuga.setheading(90)
    tortuga.fd(40)
    o=10
    while o>0:
        tortuga.write(o-1)
        tortuga.fd(50)
        o = o - 1
    return
            
def marca_jugada (tortuga, Fg, Cg, F1, C1, F2, C2):
    tortuga.penup()
    tortuga.goto(-225,-225)
    tortuga.setheading(90)
    tortuga.fd(50*9)
    tortuga.bk(50*Fg)
    tortuga.setheading(0)
    tortuga.fd(50*Cg)
    mina1voltant = Fg in {F1-1, F1, F1+1} and Cg in {C1-1, C1, C1+1}
    mina2voltant = Fg in {F2-1, F2, F2+1} and Cg in {C2-1, C2, C2+1}
    if (F1 == Fg and C1 == Cg) or (F2 == Fg and C2 == Cg):
        tortuga.color("red")
        tortuga.stamp()
        return(True)
    elif mina1voltant and mina2voltant:
        tortuga.color("red")
        tortuga.dot(15)
    elif mina1voltant | mina2voltant:
        tortuga.color("orange")
        tortuga.dot(15)
    else:
        tortuga.color("green")
        tortuga.dot(15)
    tortuga.penup()
    tortuga.goto(-250,-260)

import turtle as t
myturtle = ini_tortuga()
mywindow = ini_finestra()
ini_graella(myturtle)

import random
F1 = random.randint(0,9)
C1 = random.randint(0,9)
F2 = random.randint(0,9)
C2 = random.randint(0,9)
while F1 == F2 and C1 == C2:
    F2 = random.randint(0,9)
    C2 = random.randint(0,9)
llista_tirades = []
i=0
while i<98:
    mycellx = demana_enterx (0,9)
    mycelly = demana_entery (0,9)
    if [mycellx, mycelly] in llista_tirades:
        print("Ja havies triat aquesta cel·la abans!")
    else:
        llista_tirades.append([mycellx, mycelly])
        i+=1
        tocat = marca_jugada (myturtle, mycellx, mycelly, F1, C1, F2, C2)
        if tocat:
            myturtle.write("\tHas perdut")
            break
if i==98:
    myturtle.write("\tHas guanyat")
mywindow.exitonclick()
```
