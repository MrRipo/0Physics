## Exercici 1: ulls

He de codificar una funció que dibuixi uns ulls i que segueixi els següents requisits:
* La funció ha de tenir el nom: `dibuixar_ulls`
* Els arguments de la funció han de ser: `(tortuga,x,y,escala)` On:
 * `tortuga` és la tortuga a usar per traçar els ulls.
 * `x`,`y` són les coordenades del centre entre els dos ulls.
 * `escala` és la longitud de referència(L).

* Els centres dels ulls han d'estar situats a una distància horitzontal de L/6 respecte el punt (x,y) i el radi dels ulls ha de ser de L/10

```python
from turtle import *

finestra = Screen()
t = Turtle("turtle")
t.speed("slowest")

def dibuixar_ulls(tortuga,x,y,escala):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(0)
    tortuga.fd(escala/6)
    tortuga.setheading(270)
    tortuga.fd(escala/10)
    tortuga.setheading(0)
    tortuga.pendown()
    tortuga.circle(escala/10)
    
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(180)
    tortuga.fd(escala/6)
    tortuga.setheading(270)
    tortuga.fd(escala/10)
    tortuga.setheading(0)
    tortuga.pendown()
    tortuga.circle(escala/10)
    
dibuixar_ulls(t,0,0,500)
exitonclick()
```

## Exercici 2: nas

He de codificar una funció que dibuixi un nas i que segueixi els següents requisits:

* La funció ha de tenir el nom: `dibuixar_nas`
* Els arguments de la funció han de ser: `(tortuga,x,y,escala)` On:
 * `tortuga` és la tortuga a usar per traçar el nas.
 * `x`,`y` són les coordenades del centre del nas.
 * `escala` és la longitud de referència (L).

* El nas ha d'estar centrat a (x,y) i ha de ser un triangle de L/10 de costat.

```python
from turtle import *

finestra = Screen()
t = Turtle("turtle")
t.speed("slowest")

def dibuixar_nas(tortuga,x,y,escala):
    tortuga.penup()
    tortuga.goto(x,y)
    #tortuga.stamp() #-> Es pot afegir aquesta comanda aquí per comprovar que la marca efectivament queda en el centre del triangle.
    tortuga.setheading(90)
    tortuga.bk(0.028867513*escala) #càlcul detallat en la descripció
    tortuga.setheading(0)
    tortuga.fd(escala/20) #càlcul detallat en la descripció

    tortuga.setheading(180) 
    tortuga.pendown()
    i=0
    while i<3:
        tortuga.fd(escala/10)
        tortuga.rt(120) #Els triangles equilàters tenen tots els angles de 60º per tant gira 180-60 graus (120)
        i+=1
    
dibuixar_nas(t,0,0,5000)
    
exitonclick()
```

<span style="color:blue">La distància (x) que la tortuga ha de baixar és:</span>
$\dfrac{L/20}{sin60º} = \dfrac{x}{sin30º} → x =  \dfrac{L/20 ·sin30}{sin60} → x = 0.028867513·L$.

<span style="color:blue">La distància que la tortuga s'ha de desplaçar cap a la dreta és:</span> $\dfrac{L}{20}$.

<span style="color:blue">La primera distància la calculo utilitzant el teorema del sinus i la segona la trec de dividir per 2 la distància del costat del triangle</span> $\dfrac{\frac{L}{10}}{{2}} = \dfrac{L}{20}$ <span style="color:blue">(tal com es veu al dibuix en vermell).</span>

## Exercici 3: boca

 * `L` és la longitud de referència (escala).

```python
from turtle import *

finestra = Screen()
t = Turtle("turtle")
t.speed("slowest")

def dibuixar_boca(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(0)
    tortuga.fd(L/8)
    tortuga.pendown()
    tortuga.bk(L/4)
    
dibuixar_boca(t,0,-50,1000)
    
exitonclick()
```

##Exerici 6: Moltes cares

Ara se'm demana que usi la funció `dibuixar_cap` per escriure un programa que ompli la pantalla de 10 cares posicionades aleatòriament
i amb mides (L) aleatòries d'entre 10 i 150 punts d'escala.

```python
from turtle import *
import random

finestra = Screen()
t = Turtle("turtle")
t.speed("fastest")

def dibuixar_ulls(tortuga,x,y,escala):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(0)
    tortuga.fd(escala/6)
    tortuga.setheading(270)
    tortuga.fd(escala/10)
    tortuga.setheading(0)
    tortuga.pendown()
    tortuga.circle(escala/10)
    
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(180)
    tortuga.fd(escala/6)
    tortuga.setheading(270)
    tortuga.fd(escala/10)
    tortuga.setheading(0)
    tortuga.pendown()
    tortuga.circle(escala/10)
    
def dibuixar_nas(tortuga,x,y,escala):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(90)
    tortuga.bk(0.028867513*escala)
    tortuga.setheading(0)
    tortuga.fd(escala/20)

    tortuga.setheading(180) 
    tortuga.pendown()
    i=0
    while i<3:
        tortuga.fd(escala/10)
        tortuga.rt(120)
        i+=1
        
def dibuixar_boca(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(0)
    tortuga.fd(L/8)
    tortuga.pendown()
    tortuga.bk(L/4)
    
def dibuixar_cara(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(90)
    tortuga.bk(L/2)
    tortuga.setheading(0)
    tortuga.pendown()
    tortuga.circle(L/2)
    
def dibuixar_cap(tortuga):
    q=0
    while q<10:
        enter1 = random.randint(-300,300)
        enter2 = random.randint(-300,300)
        mides = random.randint(10,150)
        dibuixar_cara(tortuga,enter1,enter2,mides)
        dibuixar_ulls(tortuga,enter1,enter2+0.2*mides,mides)
        dibuixar_nas(tortuga,enter1,enter2,mides)
        dibuixar_boca(tortuga,enter1,enter2-0.25*mides,mides)
        q+=1

dibuixar_cap(t)
    
exitonclick()
```

```python
from turtle import *

finestra = Screen()
t = Turtle("turtle")
t.speed("slowest")

def dibuixar_cos(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(0)
    tortuga.fd(L/2)
    tortuga.setheading(270)
    tortuga.fd(L)
    tortuga.setheading(180)
    tortuga.pendown()
    j=0
    while j<2:
        tortuga.fd(L)
        tortuga.rt(90)
        tortuga.fd(2*L)
        tortuga.rt(90)
        j+=1

dibuixar_cos(t,0,0,100)

exitonclick()
```

## Millora 4: Ninot dibuixat des del nas

```python
from turtle import *

finestra = Screen()
t = Turtle("turtle")
t.speed("normal")

def Rectangle(tortuga,costat1,costat2):
    tortuga.pendown()
    r=0
    while r<2:
        tortuga.fd(costat1)
        tortuga.rt(90)
        tortuga.fd(costat2)
        tortuga.rt(90)
        r+=1
        
def dibuixar_ulls(tortuga,x,y,escala):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(90)
    tortuga.fd((escala/10)+(0.2*escala)) #Abança el radi + la distància entre ull i nas
    tortuga.setheading(0)
    tortuga.fd(escala/6) #Distància entre centre i ull dret
    tortuga.pendown()
    tortuga.circle(-escala/10)
        
    tortuga.penup()
    tortuga.bk((escala/3)) #Distància entre els dos ulls
    tortuga.pendown()
    tortuga.circle(-escala/10)
    
def dibuixar_nas(tortuga,x,y,escala):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(90)
    tortuga.bk(0.028867513*escala)
    tortuga.setheading(0)
    tortuga.fd(escala/20)

    tortuga.setheading(180) 
    tortuga.pendown()
    i=0
    while i<3:
        tortuga.fd(escala/10)
        tortuga.rt(120)
        i+=1
        
def dibuixar_boca(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(270)
    tortuga.fd(0.25*L) #Distància entre nas i boca
    tortuga.setheading(0)
    tortuga.fd(L/8)
    tortuga.pendown()
    tortuga.bk(L/4)
    
def dibuixar_cara(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(90)
    tortuga.bk(L/2)
    tortuga.setheading(0)
    tortuga.pendown()
    tortuga.circle(L/2)
    
def dibuixar_cap(tortuga,x,y,L):
    dibuixar_cara(tortuga,x,y,L)
    dibuixar_ulls(tortuga,x,y,L)
    dibuixar_nas(tortuga,x,y,L)
    dibuixar_boca(tortuga,x,y,L)
    
def dibuixar_cos(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(270)
    tortuga.fd(2.5*L) #Radi de la cara(L/2) + longitud del cos(2L)
    tortuga.rt(90)
    tortuga.fd(L/2) #Mitja amplada del cos
    tortuga.setheading(90)
    Rectangle(tortuga,2*L,L)
        
def dibuixar_bracos(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(270)
    tortuga.fd(L/2) #Radi de la cara
    tortuga.lt(90)
    tortuga.fd(L/2) #Mitja amplada del cos
    Rectangle(tortuga,L,L/8)
    tortuga.penup()
    tortuga.bk(L) #Tota l'amplada del cos
    tortuga.setheading(270)
    Rectangle(tortuga,L/8,L)
    
def dibuixar_cames(tortuga,x,y,L):
    tortuga.penup()
    tortuga.goto(x,y)
    tortuga.setheading(270)
    tortuga.fd(2.5*L) #Radi de la cara(L/2) + longitud del cos(2L)
    tortuga.lt(90)
    tortuga.fd(L/2) #Mitja amplada del cos
    tortuga.setheading(270)
    Rectangle(tortuga,L,L/8)
    tortuga.penup()
    tortuga.setheading(180)
    tortuga.fd(L-L/8) #Tota l'amplada del cos menys l'amplada de la cama
    tortuga.lt(90)
    Rectangle(tortuga,L,L/8)

def dibuixar_ninot(tortuga,x,y,L):
    dibuixar_cap(tortuga,x,y,L)
    dibuixar_cos(tortuga,x,y,L)
    dibuixar_bracos(tortuga,x,y,L)
    dibuixar_cames(tortuga,x,y,L)
    
    
dibuixar_ninot(t,50,100,120)
exitonclick()
```
