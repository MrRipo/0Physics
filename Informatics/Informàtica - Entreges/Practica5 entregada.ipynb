## Exercici 2

Opcionalmant puc estructurar el programa de manera que el dibuix del polígon estigui estructurat amb la funció:

### Dibuix del rellotge

```python
#Rellotge2
from turtle import *
speed("fast")

def rellotge(radi):
    pensize("6")
    i=0
    while i<60:
        penup()
        fd(radi)
        if i % 5 == 0:  #5=60/12
            pendown()
            fd(25)
            penup()
            bk(radi+25)
        else:
            dot(4)
            penup()
            bk(radi)
        i+=1    
        rt(360/60) 
        
rellotge(100)

exitonclick()
```

### Representació dels minuts

```python
#Busca dels minuts

from turtle import *
speed("fastest")

def rellotge(radi):
    pensize(6)
    i=0
    while i<60:
        penup()
        fd(radi)
        if i % 5 == 0:
            pendown()
            fd(25)
            penup()
            bk(radi+25)
        else:
            dot(4)
            penup()
            bk(radi)
        i+=1    
        rt(6)

        
rellotge(100)

from datetime import datetime

ara = datetime.now()

minuts = ara.minute

import turtle

finestra = turtle.Screen()
finestra.setup()
finestra.bgcolor("white")
finestra.title("Rellotge analògic")
Ripo = turtle.Turtle("turtle")
Ripo.speed ("slowest")
Ripo.setheading(0)

def busca_minuts (tortuga, minuts):
    l= 110
    minuts=minuts-15
    tortuga.pensize("10")
    tortuga.setheading((6)*(-minuts)) #360/60=6
    tortuga.fd(l)

busca_minuts(Ripo, minuts)

finestra.exitonclick()
```

### Rellotge final

```python
import turtle
from datetime import datetime


finestra = turtle.Screen()
finestra.setup()
finestra.bgcolor("white")
finestra.title("Rellotge analògic")
Ripo = turtle.Turtle("turtle")
Ripo.speed ("slowest")
Ripo.setheading(0)

ara = datetime.now()
hora = ara.hour
minuts = ara.minute

def rellotge(radi):
    Ripo2 = turtle.Turtle("arrow")
    Ripo2.speed("fastest")
    Ripo2.pensize(6)
    i=0
    while i<60:
        Ripo2.penup()
        Ripo2.fd(radi)
        if i % 5 == 0:
            Ripo2.pendown()
            Ripo2.fd(25)
            Ripo2.penup()
            Ripo2.bk(radi+25)
        else:
            Ripo2.dot(4)
            Ripo2.penup()
            Ripo2.bk(radi)
        i+=1    
        Ripo2.rt(6)
    Ripo2.penup()
    Ripo2.goto(500,500)
    
def busca_hores(tortuga, hora):
    l = 70
    if hora>12:
        hora=hora-12
    if hora == 0:
        hora=12
    hora=hora-3
    tortuga.pensize("10")
    tortuga.color("red")
    tortuga.penup()
    tortuga.goto(0,0)
    tortuga.setheading((30)*(-hora-minuts/60)) #360/12=30
    tortuga.pendown()
    tortuga.fd(l)
        
def busca_minuts(tortuga, minuts):
    l= 110
    minuts=minuts-15
    tortuga.pensize("7")
    tortuga.color("blue")
    tortuga.setheading((6)*(-minuts)) #360/60=6
    tortuga.fd(l)
    
rellotge(100)
busca_minuts(Ripo, minuts)
busca_hores(Ripo, hora)

finestra.exitonclick()
```
